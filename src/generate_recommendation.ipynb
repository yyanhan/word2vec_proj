```python
input_text = "./words.txt"
output = "./query_result.txt"

inp = open(input_text, 'r')
oup = open(output, 'w')

filtered_query = []
result_single = []
result_permutation = []
result_positive = []
amount
```

# query preprocessing 

1. in lower case
2. only noun, verb pos

```python
for query in inp.readlines():
	words = query.split()
	#print(TextBlob(query).tags)

	filtered_word = []
	for w in words:	# each word in a query
		# w = Word(w).singularize()
		# w = Word(w).lemmatize()
		w = Word(w).lower()
		pos = nltk.pos_tag([w]) # [('music', 'NN')]
		#print(pos)
		if not pos[0][1].find("NN") or not pos[0][1].find("JJ") or not pos[0][1].find("VB"):
			filtered_word.append(str(w))
			# print(pos)
		
	filtered_query.append(filtered_word)

print(filtered_query)
# inp.close()
amount = len(filtered_query)
```

```python
def getsimilar(word,n):
	"""get most similar n words

	Args:
		word (str): target word
		n (int): top n words

	Returns:
		list: most n similar words of word
	"""
	r = model.wv.most_similar(word)
	return r[0:n]
```

```python
for query in filtered_query:
	result_word = []
	for word in query:
		r = getsimilar(word,3)
		# r = model.wv.most_similar(word)
		# r = model.most_similar(word)
		print(r)
		result_word.append(r)
	result_single.append(result_word)
print("=====")
```

```python
def getpermutation(lst): # each query, not all query
	"""replace keywords by similar words
	"""
	# print(lst)
	result = []
	for i in product(*lst):	#each permutation	A,A,B
		line = ""
		score = 0
		for w in i:	# each word conbine a query
			line = line + " " + w[0]	
			score = w[1] + score
		line = line.lstrip()
		#print(score/len(i))
		result.append((line,score/len(i)))
	#print(result)
	r = sorted(set(result), key=lambda x:x[1],reverse=True) 
	#print(r)
	return r
```

```python
for single in result_single:
	result_permutation.append(getpermutation(single))
	#getpermutation(single)
print(result_permutation)
```

```python
def getpositive(word,n):
	r = model.wv.most_similar(positive=word)
	r = sorted(set(r), key=lambda x:x[1],reverse=True) 
	return r[0:n]
```

```python
for query in filtered_query:
	r = getpositive(query,3)
	result_positive.append(r)
print(result_positive)
```

```python
# combine permutation and positive
result = []
for i in range (0,len(filtered_query)):
    r = result_permutation[i] + result_positive[i]
    r = sorted(set(r), key=lambda x:x[1],reverse=True) 
    result.append(r)
for r in result:
    print(r)
```
